```python
with open("data/FAO_AS_4537.csv", "r") as f:
    data = pd.read_csv(f)

data.info()
```

```python
selected_data = data[["FREQ", "FREQ_LABEL", "REF_AREA", "REF_AREA_LABEL", "SEX", "SEX_LABEL", "AGE", "AGE_LABEL", "URBANISATION", "URBANISATION_LABEL", "TIME_PERIOD","OBS_VALUE"]]
selected_data.head()
```

URBANISATION: Is always total, """"
AGE: Is always total, " " "
